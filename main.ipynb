```python
import pandas as pd
import folium
```

```python
brasil = folium.Map(
    location=[-16.1237611, -59.9219642],
    zoom_start=4
)
```

```python
brasil
```

```python
empresas = pd.read_csv('empresas.xz')
```

```python
empresa1 = empresas.iloc[0]
folium.Marker(
    location=[empresa1['latitude'], empresa1['longitude']],
).add_to(brasil)
```

```python
empresa2 = empresas.iloc[1]
folium.Marker(
    location=[empresa2['latitude'], empresa2['longitude']],
).add_to(brasil)
```

```python
brasil
```

```python
empresas_mt = empresas[empresas['state'] == 'MT']
```

```python
matogrosso = folium.Map(
    location=[-12.6470202, -55.4237152],
    zoom_start=6
)
```

```python
matogrosso
```

```python
for _, empresa in empresas_mt.iterrows():
    folium.Marker(
        location=[empresa['latitude'], empresa['longitude']],
    ).add_to(matogrosso)
```

```python
matogrosso
```

```python
empresas_cuiaba_mt = empresas[empresas['city'] == 'CUIABA']
```

```python
cuiaba = folium.Map(
    location=[-15.5989, -56.0949],
    zoom_start=12
)
```

```python
for _, empresa in empresas_cuiaba_mt.iterrows():
    folium.Marker(
        location=[empresa['latitude'], empresa['longitude']],
    ).add_to(cuiaba)
```

```python
cuiaba
```

```python
colors = {
    'DOM AQUINO': 'pink',
    'BOSQUE DA SAUDE': 'blue',
    'LIXEIRA': 'gree',
    'MORADA DO OURO': 'orange',
    'CENTRO SUL': 'red',
    'CPA': 'purple'
}
```

```python
cuiaba = folium.Map(
    location=[-15.5989, -56.0949],
    zoom_start=12
)
```

```python
for _, empresa in empresas_cuiaba_mt.iterrows():
    if empresa['neighborhood'] in colors.keys():
        folium.Marker(
            location=[empresa['latitude'], empresa['longitude']],
            icon=folium.Icon(color=colors[empresa]['neighborhood'])
        ).add_to(cuiaba)
```
